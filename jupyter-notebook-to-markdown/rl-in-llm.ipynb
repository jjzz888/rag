### Multi-reward Setting

A multi-reward setting is a reinforcement learning setup where each action/trajectory is evaluated by multiple reward functions at the same time, instead of a single scalar reward.

Formally, instead of one reward $r$, you have a reward vector:

$$\mathbf{r} = (r_1, r_2, \dots, r_K)$$

Each $r_k$ corresponds to a different objective.

#### Why does this matter for LLM RL?

For language models, “good behavior” is never one-dimensional.

Examples of objectives you often want simultaneously:

| Reward      | What it measures                             |
| ----------- | -------------------------------------------- |
| Correctness | Is the answer factually / logically correct? |
| Format      | Does it follow the required output format?   |
| Length      | Is it concise or within a token budget?      |
| Safety      | Does it avoid unsafe content?                |
| Style       | Politeness, tone, helpfulness                |

So for one generated response $y$, you might compute:

$$
\mathbf{r}(y) = (\text{accuracy}, \text{format}, \text{length})
$$

That’s a multi-reward signal.

#### How multi-reward RL is usually done (naively)

The standard approach is to collapse rewards into a scalar:

$$
r_{\text{total}} = \sum_{k=1}^{K} w_k , r_k
$$

Then just run PPO / GRPO as usual. This hides trade-offs:

* A response that is very correct but poorly formatted.
* vs. one that is perfectly formatted but wrong.

Both can end up with the same total reward. Once collapsed, the optimizer can’t tell why a response is good or bad.

#### Multi-reward with group-based RL (GRPO setup)

In GRPO-style RLHF:

* For each prompt $x_i$
* Sample a group of responses $({y_{i,1}, \dots, y_{i,G}})$
* Compute rewards relative to others in the group. This avoids needing a learned value function.

In a multi-reward case, each response has:
$$
\mathbf{r}_{i,j} = (r_{i,j}^{(1)}, \dots, r_{i,j}^{(K)})
$$

The key question becomes how to compute advantages from this reward vector.
